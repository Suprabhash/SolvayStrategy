```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



# Data for the table
data = {
    "Company": ["Zara/Inditex", "Asos", "H&M", "Gap", "Primark", "Ross Stores", "Benetton", "Boohoo", "Shein"],
    "Digital Presence": [2.8, 2.9, 2.7, 1.5, 1.2, 1.1, 1.3, 2.6, 2.9],
    "Fashionability": [2.9, 2.7, 2.8, 1.5, 1.3, 1.2, 1.4, 2.5, 2.6],
    "Affordability": [2.1, 1.5, 2.3, 2.0, 1.1, 1.2, 1.9, 1.4, 1.3]
}

# Creating the DataFrame
df = pd.DataFrame(data)

x = df['Digital Presence'].to_numpy()
y = df['Fashionability'].to_numpy()
z = df['Affordability'].to_numpy()


# Creating figure for combined plot
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# Plotting your scatter plot
# ax.scatter3D(x, y, z, color="green", label='Random Data')

# Plotting from DataFrame
for i in range(len(df)):
    ax.scatter3D(df['Digital Presence'][i], df['Fashionability'][i], df['Affordability'][i], label=df['Company'][i])

# Adding labels and title
ax.set_xlabel('Digital Presence')
ax.set_ylabel('Fashionability')
ax.set_zlabel('Affordability')
plt.title("3D Scatter Plot with Company Data")

# Adding legend
ax.legend()
# Removing ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Removing gridlines
ax.grid(False)
plt.xlim([0, 3])
plt.ylim([0, 3])
# Show plot
plt.show()
```
